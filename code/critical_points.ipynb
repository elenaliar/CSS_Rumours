Notebook to check how the model behaves for different pairs  of density/spread threshold

```python
from visualisation import plot_percolation_vs_density_vs_spread_threshold, plot_3d_percolation_vs_density_and_threshold, plot_3d_gossip_spreader_counts, plot_time_status, plot_log_log_distribution, simulate_and_plot_gossip_model_all_combinations
import matplotlib.pyplot as plt

densities = [0.6, 0.7, 0.8]
spread_thresholds = [0.1, 0.2, 0.3, 0.4]
grid_size = 20
num_simulation = 500
steps = 100
x_limits = (0, steps-20)
y_limits = (0, grid_size*grid_size)


#Moore neighborhood
print("Simulations for Moore neighborhood")
#flag_center=1 by default 
fig, axs = plt.subplots(3, 4, figsize=(14, 12), constrained_layout=True)
fig.suptitle("Status counts over time", fontsize=24, y=1.03)
index = 0
flag_center=1
for i, density in enumerate(densities):
    for j, spread_threshold in enumerate(spread_thresholds):
        ax = axs[i, j]
        plot_time_status(ax, grid_size, density, spread_threshold, steps, num_simulation, flag_center, x_limits, y_limits)
        index += 1

handles, labels = axs[0, 0].get_legend_handles_labels()
fig.legend(handles, labels, loc='upper center', fontsize=14, ncol=4, bbox_to_anchor=(0.5, 1.05), frameon= False)
plt.show()

plot_3d_percolation_vs_density_and_threshold(grid_size, steps, num_simulation)
plot_3d_gossip_spreader_counts(grid_size, steps, num_simulation)

plot_percolation_vs_density_vs_spread_threshold(grid_size, steps, num_simulation)

#flag_center=0
fig, axs = plt.subplots(3, 4, figsize=(14, 12), constrained_layout=True)
fig.suptitle("Status counts over time", fontsize=24, y=1.03)
index = 0
flag_center=0   
for i, density in enumerate(densities):
    for j, spread_threshold in enumerate(spread_thresholds):
        ax = axs[i, j]
        plot_time_status(ax, grid_size, density, spread_threshold, steps, num_simulation, flag_center, x_limits, y_limits)
        index+=1

handles, labels = axs[0, 0].get_legend_handles_labels()
fig.legend(handles, labels, loc='upper center', fontsize=14, ncol=4, bbox_to_anchor=(0.5, 1.05), frameon= False)
plt.show()

plot_3d_percolation_vs_density_and_threshold(grid_size, steps, num_simulation, flag_center)
plot_3d_gossip_spreader_counts(grid_size, steps, num_simulation, flag_center)
  
plot_percolation_vs_density_vs_spread_threshold(grid_size, steps, num_simulation, flag_center)

#von Neumann neighborhood
print("Simulations for Von Neumann neighborhood")
flag_neighbors=0
#flag_center=1
flag_center=1
plot_percolation_vs_density_vs_spread_threshold(grid_size, steps, num_simulation, flag_center, flag_neighbors)

#flag_center=0
flag_center=0
plot_percolation_vs_density_vs_spread_threshold(grid_size, steps, num_simulation, flag_center, flag_neighbors)
```
