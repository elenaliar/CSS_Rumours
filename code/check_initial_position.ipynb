### Ways to show how the initial position affects the percolation

```python
import numpy as np 
from simulation import run_multiple_simulations_for_phase_diagram, run_simulation, count_statuses
from visualisation import show_lecture_hall_over_time, plot_time_status
from ca_lecture_hall_model import Grid
from matplotlib import pyplot as plt

size = 20
spread_threshold = [0.1, 0.2, 0.3]
densities = [0.4, 0.5, 0.6]
num_simulations = 500
steps = 100

#put initial spreader in center
flag_center = 0
count_percolation_center = 0
count_percolation_edge = 0
for spread in spread_threshold:
    for density in densities:
        #initial spreader in center
        
        print(f"Running simulation for density {density} and spread {spread} and initial spreader in center")
        g_c = Grid(size, density, spread)
        g_c.initialize_board()
        grids = run_simulation(g_c, steps)
        show_lecture_hall_over_time(grids)
        if g_c.check_percolation():
            count_percolation_center += 1
        plt.show()

        #initial spreader outside of center
        print(f"Running simulation for density {density} and spread {spread} and initial spreader outside of center")
        g_e = Grid(size, density, spread)
        g_e.initialize_board(flag_center)
        grids = run_simulation(g_e, steps)
        show_lecture_hall_over_time(grids)
        if g_e.check_percolation():
            count_percolation_edge += 1
        plt.show()

print(f"Percolation rate for center: {count_percolation_center/(len(spread_threshold)*len(densities))}")
print(f"Percolation rate for edge: {count_percolation_edge/(len(spread_threshold)*len(densities))}")


```
